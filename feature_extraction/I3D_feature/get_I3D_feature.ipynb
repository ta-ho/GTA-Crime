```python
from tqdm.auto import tqdm
import os
import glob
import numpy as np

# 영상 저장 경로 (안에 카테고리별로 영상 폴더를 만들어줘야 함)
root_path = "D:/newGTAVideo"
folders = os.listdir(root_path)
print(folders)
# 피처 저장 경로
save_path = "D:/newGTAI3D"
os.makedirs(save_path, exist_ok=True)
```

```python
from models.i3d.extract_i3d import ExtractI3D
from utils.utils import build_cfg_path
from omegaconf import OmegaConf
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)

feature_type = 'i3d'

args = OmegaConf.load(build_cfg_path(feature_type))
```

```python
for folder in tqdm(folders):
  print(folder)
  video_path = glob.glob(os.path.join(os.path.join(root_path, folder), "*.mp4"))
  args.video_paths = video_path
  extractor = ExtractI3D(args)
  
  os.makedirs(os.path.join(save_path, folder), exist_ok=True)
  
  for video in tqdm(video_path):
    npy_feature = extractor.extract(video)['rgb'].squeeze(2)
    print(npy_feature.shape)
    for i in range(10):
      np.save(os.path.join(os.path.join(save_path, folder), video.split('\\')[-1].split('.')[0] + "__" + str(i) + ".npy".format(i)), npy_feature[:, i, :])
```
